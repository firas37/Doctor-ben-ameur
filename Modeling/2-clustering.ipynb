```python
from sklearn.cluster import KMeans
wcss = [] 
max=20
for i in range(1, max): 
    kmeans = KMeans(n_init = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X) 
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, max), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
from sklearn.cluster import KMeans
n_clusters = 7
kmeans = KMeans(n_init=n_clusters, init="k-means++", random_state=42)
kmeans.fit(matrix)
labels = kmeans.labels_
df["Cluster"] = labels
```

```python
# Centroid coordinates
centroids = kmeans.cluster_centers_
```

```python
matrix.shape
```

```python
import scipy.spatial.distance as sdist
```

```python
centroids.shape
```

```python
centroids[df['Cluster']].shape
```

```python
def distance(index):
    dist=sdist.norm(matrix[index] - centroids[df['Cluster']][index])
    return dist
```

```python

dist=[distance(index) for index in range(len(df['Cluster']))]
```

```python
df['distance']=dist
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

tsne = TSNE(n_components=2, perplexity=15, random_state=42, init="random", learning_rate=200)
vis_dims2 = tsne.fit_transform(matrix)

# Reverse X and Y axes by swapping the components
y = [x for x, y in vis_dims2]  # First component becomes Y-axis
x = [y for x, y in vis_dims2]  # Second component becomes X-axis

plt.figure(figsize=(10, 6))

for category, color in enumerate(["purple", "green", "red", "blue", "yellow", "orange", "violet", "cyan", "magenta"]):
    xs = np.array(x)[df.Cluster == category]
    ys = np.array(y)[df.Cluster == category]
    
    plt.scatter(xs, ys, color=color, alpha=0.3, label=f'Cluster {category}')

    # Calculate centroid (using reversed axes)
    avg_x = xs.mean()
    avg_y = ys.mean()

    plt.scatter(avg_x, avg_y, marker="X", color=color, s=100, edgecolor='black')

plt.title("Clusters visualized in 2D using t-SNE (Axes Reversed)")
plt.xlabel("t-SNE Y-axis (original X)")  # Swapped labels
plt.ylabel("t-SNE X-axis (original Y)")  # Swapped labels
plt.legend()
plt.show()
```
